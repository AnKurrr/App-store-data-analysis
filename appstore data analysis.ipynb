## cleaning data ##

```python
#as we don't need some columns so we are removing them
data.drop(["Unnamed: 0","id","currency","rating_count_ver","user_rating_ver","ver","sup_devices.num","ipadSc_urls.num","lang.num","vpp_lic"],axis=1,inplace=True)
```

```python
data.head()
```

```python
# size_bytes column is having value in bytes converting that value in mb in different column
data['size_mb'] = data['size_bytes'] / (1024 * 1024)
data = data.drop(columns=['size_bytes']) # Droping  the 'size_bytes' column
```

```python
data.head(2)
```

## data analysis ##

```python
#data cleaning is done so lets start some data analysis
```

```python
#finding total number of apps in playstore 
x=data.track_name.count()
print("Total Number of Apps in App Store Are",x)
```

```python
#finding total genres and types of apps in App store and the total number of apps in that genre
x=data.groupby(['prime_genre'])['track_name'].count()
print(x)
```

```python
#percentage of each genere in app store
data['prime_genre'].value_counts(normalize=True) * 100
```

```python
new_data_cate=data.groupby([data.prime_genre])[['track_name']].count().reset_index().sort_values('track_name' ,ascending = False)
new_data_cate.columns = ['prime_genre','# of Apps']
new_data_cate.head()
```

```python
sns.barplot(y = 'prime_genre',x = '# of Apps', data=new_data_cate.head(10))
```

```python
#Top 10 most downloaded or rated free app
data[data['price'] == 0].sort_values('rating_count_tot', ascending=False).head(10)
```

```python
#Top 10 highest rated paid apps
data[data['price'] != 0].sort_values('rating_count_tot', ascending=False).head(10)
```

```python
#10 biggest size app in app store 
data.sort_values('size_mb',ascending=False).head(10)
```

```python
#total number of app that are free on app store 
x=len(data[data['price'] == 0])
print("Total Number Of App Tat Are Free ON App Store Are",x)
```

```python
x=len(data[data['price'] != 0])
print("Total Number Of App Tat Are Paid ON App Store Are",x)
```

```python
#average user rating of each category are
x=data.groupby(["prime_genre"]).agg({"prime_genre":"size","user_rating":"mean"})
x
```

```python
#checking app price distribution
plt.figure(figsize=(12,4))
sns.distplot(data["price"],bins=100,rug=True,color="Purple")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.title("App Price Distribution")
plt.savefig("/Users/ankur/Downloads/app_price_distribution.png", dpi=300, bbox_inches='tight')
```

```python
#most expensive app on Appstore 
x = data.sort_values(by='price', ascending=False).iloc[0]['track_name']
print("The Most Expensive App on App  Store Is ",x)
```

```python
#chwcking if apps are safe for childeren or not 
sns.countplot(x=data["cont_rating"], order=data["cont_rating"].value_counts().index, palette="YlOrRd")
plt.xlabel("Content Rating")
plt.ylabel("Count")
plt.savefig("/Users/ankur/Downloads/content.png", dpi=300, bbox_inches='tight')
```

```python
#Top 10 Games On App Store To PLay As per the Ratings Are 
data[data['prime_genre'] == 'Games'].sort_values('rating_count_tot', ascending=False).head(10)
```

```python
# A histogram showing the relation between user ratings and an app category
sns.histplot(data=data, x='user_rating', hue='prime_genre', multiple='stack', bins=7)
```
